#### Consuming APIs

```python
import requests
import json

response = requests.get('https://api.stackexchange.com/2.3/answers?order=desc&sort=activity&site=stackoverflow')
```

```python
response.json()
```

```python
response.json()['items']
```

```python
for data in response.json()['items']:
    print(data['owner']['display_name'])
    print(data['owner']['link'])
    print()
```
